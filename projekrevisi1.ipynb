```python
# %% [markdown]
# # Analisis Tingkat Kemiskinan di Indonesia

# %% [markdown]
# ## 1. Import Library

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, 
                             confusion_matrix, 
                             accuracy_score,
                             RocCurveDisplay,
                             PrecisionRecallDisplay)
from imblearn.over_sampling import SMOTE
import joblib

# %% [markdown]
# ## 2. Load dan Pembersihan Data

# %%
# Membaca data dengan separator titik koma
df = pd.read_csv('Klasifikasi Tingkat Kemiskinan di Indonesia.csv', sep=';')

# Membersihkan data
df = df.dropna(how='all')  # Hapus baris yang seluruhnya kosong
df = df.replace(',', '.', regex=True)  # Ganti koma dengan titik untuk desimal

# Kolom numerik yang akan digunakan
numeric_cols = [
    'Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)',
    'Rata-rata Lama Sekolah Penduduk 15+ (Tahun)',
    'Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)',
    'Indeks Pembangunan Manusia',
    'Umur Harapan Hidup (Tahun)',
    'Persentase rumah tangga yang memiliki akses terhadap sanitasi layak',
    'Persentase rumah tangga yang memiliki akses terhadap air minum layak',
    'Tingkat Pengangguran Terbuka',
    'Tingkat Partisipasi Angkatan Kerja',
    'PDRB atas Dasar Harga Konstan menurut Pengeluaran (Rupiah)'
]

# Konversi ke tipe data numerik
for col in numeric_cols:
    # Menghapus titik sebagai pemisah ribuan dan mengubah koma menjadi titik untuk desimal
    df[col] = df[col].astype(str).str.replace('.', '', regex=False)
    df[col] = df[col].str.replace(',', '.', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Hapus baris dengan nilai kosong di kolom numerik
df = df.dropna(subset=numeric_cols)

# Kolom target
target_col = 'Klasifikasi Kemiskinan'

# %% [markdown]
# ## 3. Eksplorasi Data

# %%
print(f"\nJumlah baris: {df.shape[0]}, Jumlah kolom: {df.shape[1]}")
print("\n5 data teratas:")
display(df.head())

# %%
print("\nStatistik Deskriptif:")
display(df[numeric_cols].describe().T)

# %%
# Distribusi klasifikasi kemiskinan
plt.figure(figsize=(8, 6))
sns.countplot(x=target_col, data=df)
plt.title('Distribusi Klasifikasi Kemiskinan')
plt.xlabel('Klasifikasi (0: Tidak Miskin, 1: Miskin)')
plt.ylabel('Jumlah Kabupaten/Kota')
plt.show()

# Pilih kolom numerik yang ingin divisualisasikan
selected_columns = [
    'Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)',
    'Rata-rata Lama Sekolah Penduduk 15+ (Tahun)',
    'Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)',
    'Indeks Pembangunan Manusia',
    'Umur Harapan Hidup (Tahun)',
    'Persentase rumah tangga yang memiliki akses terhadap sanitasi layak',
    'Persentase rumah tangga yang memiliki akses terhadap air minum layak',
    'Tingkat Pengangguran Terbuka',
    'Tingkat Partisipasi Angkatan Kerja',
    'PDRB atas Dasar Harga Konstan menurut Pengeluaran (Rupiah)'
]

# Buat grid plot
plt.figure(figsize=(20, 15))
for i, col in enumerate(selected_columns, 1):
    plt.subplot(4, 3, i)
    sns.histplot(df[col], kde=True, bins=30, color='steelblue', edgecolor='black')
    plt.title(col, fontsize=10)
    plt.xlabel(col, fontsize=9)
    plt.ylabel('Jumlah daerah (kabupaten/kota)', fontsize=9)
    plt.xticks(rotation=15)
    plt.tight_layout()

plt.suptitle('Distribusi Variabel Sosial Ekonomi di Indonesia', fontsize=14, y=1.02)
plt.subplots_adjust(top=0.92)
plt.show()

# Matriks korelasi
plt.figure(figsize=(12, 10))
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Korelasi Antar Variabel Numerik')
plt.show()

# %% [markdown]
# ## 4. Visualisasi Kunci

# %%
# Distribusi pengeluaran per kapita
plt.figure(figsize=(10, 6))
sns.boxplot(x=target_col, 
            y='Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)', 
            data=df)
plt.title('Distribusi Pengeluaran per Kapita berdasarkan Klasifikasi Kemiskinan')
plt.show()

# %%
# Hubungan pendidikan dan kemiskinan
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Rata-rata Lama Sekolah Penduduk 15+ (Tahun)',
    y='Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)',
    hue=target_col,
    data=df,
    alpha=0.7
)
plt.title('Hubungan Pendidikan dan Tingkat Kemiskinan')
plt.show()

# %%
# 10 daerah termiskin
top_miskin = df.sort_values('Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)', 
                          ascending=False).head(10)
plt.figure(figsize=(12, 8))
sns.barplot(
    x='Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)',
    y='Kab/Kota',
    data=top_miskin,
    palette='Reds_d'
)
plt.title('10 Kabupaten/Kota dengan Tingkat Kemiskinan Tertinggi')
plt.show()

# %% [markdown]
# ## 5. Pemodelan Klasifikasi

# %%
# Persiapan data
X = df[numeric_cols]
y = df[target_col]

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Normalisasi
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Penanganan data tidak seimbang
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

# %%
# Pelatihan model dengan GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid, 
    cv=5, 
    scoring='f1',  # Menggunakan f1 karena data tidak seimbang
    n_jobs=-1
)
grid_search.fit(X_train_smote, y_train_smote)

best_model = grid_search.best_estimator_
print("Parameter terbaik:", grid_search.best_params_)

# %%
# Evaluasi model
y_pred = best_model.predict(X_test_scaled)

print("\nAkurasi:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=['Tidak Miskin', 'Miskin'],
            yticklabels=['Tidak Miskin', 'Miskin'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve
RocCurveDisplay.from_estimator(best_model, X_test_scaled, y_test)
plt.title('ROC Curve')
plt.show()

# Precision-Recall Curve
PrecisionRecallDisplay.from_estimator(best_model, X_test_scaled, y_test)
plt.title('Precision-Recall Curve')
plt.show()

# %% [markdown]
# ## 6. Feature Importance

# %%
# Analisis kepentingan fitur
feature_importance = pd.Series(best_model.feature_importances_, index=numeric_cols)
feature_importance_df = feature_importance.sort_values(ascending=False).reset_index()
feature_importance_df.columns = ['Fitur', 'Skor Importance']

plt.figure(figsize=(10, 6))
sns.barplot(x='Skor Importance', y='Fitur', data=feature_importance_df, palette='viridis')
plt.title('Kepentingan Fitur dalam Model Random Forest')
plt.xlabel('Skor Kepentingan')
plt.ylabel('Fitur')
plt.tight_layout()
plt.show()

# Simpan hasil analisis fitur
feature_importance_df.to_csv('feature_importance_kemiskinan.csv', index=False)

# %% [markdown]
# ## 7. Simpan Hasil Analisis

# %%
# Simpan prediksi dan model
df['Prediksi'] = best_model.predict(scaler.transform(X))
df.to_csv('hasil_klasifikasi_kemiskinan.csv', index=False)

joblib.dump(best_model, 'model_kemiskinan.pkl')
joblib.dump(scaler, 'scaler_kemiskinan.pkl')

print("\nAnalisis selesai. Model, scaler, dan hasil prediksi telah disimpan.")
```
